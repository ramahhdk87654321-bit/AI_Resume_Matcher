```python
%%writefile app.py

import streamlit as st
import pdfplumber
import pandas as pd
import ast

from sentence_transformers import SentenceTransformer, util

# -----------------------------
# Page Config
# -----------------------------

st.set_page_config(
    page_title="AI Resume Matcher",
    page_icon="💼",
    layout="wide"
)

# -----------------------------
# Custom CSS
# -----------------------------

st.markdown("""
<style>

.main{
    background-color:#F5F7FB;
}

.big-title{
    font-size:42px;
    font-weight:bold;
    color:#2E3A59;
    text-align:center;
}

.subtitle{
    text-align:center;
    color:gray;
    font-size:18px;
}

.result{
    background:#ffffff;
    padding:20px;
    border-radius:15px;
    box-shadow:0px 3px 12px rgba(0,0,0,0.1);
}

</style>
""",unsafe_allow_html=True)

st.markdown("<p class='big-title'>💼 AI Resume Matcher</p>",unsafe_allow_html=True)

st.markdown(
"<p class='subtitle'>Upload your CV and discover the most suitable job.</p>",
unsafe_allow_html=True
)

# -----------------------------
# Load Model
# -----------------------------

@st.cache_resource
def load_model():
    return SentenceTransformer("all-MiniLM-L6-v2")

model = load_model()

# -----------------------------
# Load Dataset
# -----------------------------

@st.cache_data
def load_jobs():

    jobs = pd.read_csv("all_job_post.csv")

    jobs = jobs[['category',
                 'job_title',
                 'job_description',
                 'job_skill_set']]

    jobs.dropna(inplace=True)

    jobs.drop_duplicates(inplace=True)

    jobs["job_skill_set"] = jobs["job_skill_set"].apply(ast.literal_eval)

    return jobs

jobs = load_jobs()

# -----------------------------
# Upload CV
# -----------------------------

uploaded_file = st.file_uploader(
    "📄 Upload Your CV",
    type="pdf"
)

# -----------------------------
# Read PDF
# -----------------------------

def read_pdf(pdf):

    text=""

    with pdfplumber.open(pdf) as pdf_file:

        for page in pdf_file.pages:

            if page.extract_text():

                text += page.extract_text()

    return text

# -----------------------------
# Prediction
# -----------------------------

if uploaded_file:

    cv_text = read_pdf(uploaded_file)

    with st.spinner("Analyzing Resume..."):

        cv_embedding = model.encode(cv_text,
                                    convert_to_tensor=True)

        job_embeddings = model.encode(
            jobs["job_description"].tolist(),
            convert_to_tensor=True
        )

        similarities = util.cos_sim(
            cv_embedding,
            job_embeddings
        )[0]

        best_index = similarities.argmax().item()

        best_job = jobs.iloc[best_index]

        score = similarities[best_index].item()*100

    st.success("Analysis Finished ✅")

    st.markdown("---")

    col1,col2=st.columns(2)

    with col1:

        st.metric(
            " Match Score",
            f"{score:.2f}%"
        )

    with col2:

        st.metric(
            "💼 Best Job",
            best_job["job_title"]
        )

    st.markdown("---")

    st.subheader("📋 Job Description")

    st.write(best_job["job_description"])

    st.markdown("---")

    job_skills = [
        s.lower()
        for s in best_job["job_skill_set"]
    ]

    cv_lower = cv_text.lower()

    matched=[]

    missing=[]

    for skill in job_skills:

        if skill in cv_lower:

            matched.append(skill)

        else:

            missing.append(skill)

    skills_score=(len(matched)/len(job_skills))*100

    st.progress(skills_score/100)

    st.write(
        f"### Skills Match : {skills_score:.2f}%"
    )

    c1,c2=st.columns(2)

    with c1:

        st.success("✅ Matched Skills")

        st.write(matched)

    with c2:

        st.error("❌ Missing Skills")

        st.write(missing)
```
